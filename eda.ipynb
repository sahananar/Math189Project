# EDA Question one: Is Missingness dependent on airline reporting?

```python
missingness_by_airline = df.groupby("Reporting_Airline").apply(lambda x: x.isnull().sum(axis=0)/x.shape[0])
missingness_by_airline.to_csv("missingness.csv")
missingness_by_airline
```

Looks like there are some diffrences by airline, one didn't report having any delays (?)
